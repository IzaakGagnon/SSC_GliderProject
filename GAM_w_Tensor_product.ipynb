```python
import io
import h5py
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from pygam import LinearGAM, s, te
from sklearn.metrics import mean_squared_error

# unpack H5 files
def load_h5_as_df(file_bytes):
    with h5py.File(io.BytesIO(file_bytes), 'r') as f:
        data = f['data/block0_values'][:]
        colnames = [name.decode('utf-8') for name in f['data/block0_items'][:]]
    return pd.DataFrame(data, columns=colnames)

dfs = []
for filename, file_bytes in uploaded.items():
    if filename.endswith(".h5"):
        print(f"Processing {filename} ...")
        df_temp = load_h5_as_df(file_bytes)
        dfs.append(df_temp)

if dfs:
    data_all = pd.concat(dfs, ignore_index=True)
else:
    raise ValueError("No HDF5 files found among the uploaded files.")

# rename target variable as oxygen
data_all = data_all.rename(columns={'oxygen': 'micromoles_of_oxygen_per_unit_mass_in_sea_water'})
data_all = data_all.dropna()

print("Combined data preview:")
print(data_all.head())

# divide depth into quantiles
depth_30 = data_all['depth'].quantile(0.3)
depth_35 = data_all['depth'].quantile(0.35)
depth_65 = data_all['depth'].quantile(0.65)
depth_70 = data_all['depth'].quantile(0.7)
print("Depth thresholds (20%, 40%, 60%, 80%):", depth_30, depth_35, depth_65, depth_70)

# split data into train test and run GAM model with tensor product for long/latitudes
def run_gam_split(train_data, test_data, label=""):
    print("\n=============================")
    print("Evaluating:", label)
    print("=============================")

    # 3D Scatter Plot
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(train_data['longitude'], train_data['latitude'], -train_data['depth'],
               c='grey', s=1, label='Train')
    ax.scatter(test_data['longitude'], test_data['latitude'], -test_data['depth'],
               c='red', s=1, label='Test')
    ax.set_xlabel("Longitude")
    ax.set_ylabel("Latitude")
    ax.set_zlabel("Depth (negative)")
    ax.set_title(f"Train/Test Split: {label}")
    plt.legend()
    plt.show()

    # GAM
    X_train = train_data[['latitude', 'longitude', 'depth', 'temperature', 'salinity']]
    y_train = train_data['micromoles_of_oxygen_per_unit_mass_in_sea_water']

    X_test = test_data[['latitude', 'longitude', 'depth', 'temperature', 'salinity']]
    y_test = test_data['micromoles_of_oxygen_per_unit_mass_in_sea_water']

    gam = LinearGAM(te(0, 1) + s(2) + s(3) + s(4)).fit(X_train, y_train)
    preds = gam.predict(X_test)

    # evaluation metrics
    mse = mean_squared_error(y_test, preds)
    r2 = gam.score(X_test, y_test)

    print(f"MSE: {round(mse, 2)}")
    print(f"R²: {round(r2, 3)}")
    print("\nModel Summary:")
    print(gam.summary())

# evaluate specific portions of the data

# shallow: test points from shallowest 20% (depth <= depth_30)
shallow_data = data_all[data_all['depth'] <= depth_30]
if len(shallow_data) >= 100000:
    test_shallow = shallow_data.sample(n=100000, random_state=42)
else:
    test_shallow = shallow_data.sample(frac=0.5, random_state=42)
train_shallow = data_all.drop(index=test_shallow.index)
print("Shallow test depth range:", test_shallow['depth'].min(), "to", test_shallow['depth'].max())
run_gam_split(train_shallow, test_shallow, "Shallowest Depth Band")

# middle: test points from 40–60% depth (depth between depth_35 and depth_65)
middle_data = data_all[(data_all['depth'] >= depth_35) & (data_all['depth'] <= depth_65)]
if len(middle_data) >= 100000:
    test_middle = middle_data.sample(n=100000, random_state=43)
else:
    test_middle = middle_data.sample(frac=0.5, random_state=43)
train_middle = data_all.drop(index=test_middle.index)
run_gam_split(train_middle, test_middle, "Middle Depth Band")

# deep: test points from deepest 20% (depth >= depth_70)
deep_data = data_all[data_all['depth'] >= depth_70]
if len(deep_data) >= 100000:
    test_deep = deep_data.sample(n=100000, random_state=44)
else:
    test_deep = deep_data.sample(frac=0.5, random_state=44)
train_deep = data_all.drop(index=test_deep.index)
run_gam_split(train_deep, test_deep, "Deepest Depth Band")
```
